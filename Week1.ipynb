```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://upload.wikimedia.org/wikipedia/commons/thumb/9/9c/1-month-old_kitten_45.jpg/800px-1-month-old_kitten_45.jpg"]

for url in urls:
        image = io.imread(url) #read image
        image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
        image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) #convert color to RGB
        final_frame = cv.hconcat((image, image_2)) #concatenate image
        cv2_imshow(final_frame) #show image
        print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis vertical ditengah image
for x in range (tinggi):
  image_4[x,int((lebar)/2)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_4))
cv2_imshow(final_frame)

for x in range (tinggi):
  y = int(x * (lebar/tinggi))   #Mengatur nilai y berdasarkan perbandingan lebar dan tinggi
  z = lebar - y - 1             #Mengatur nilai z berdasarkan y
  image_5[x, y] = [255,255,255]
  image_5[x, z] = [255,255,255]

final_frame = cv.hconcat((image_2, image_5))
cv2_imshow(final_frame)
```

5. Jelaskan keterkaitan antara pixel dan juga resolusi gambar yang tinggi ataupun rendah!

- Jawab: Keterkaitan antara pixel dan resolusi yaitu resolusi gambar (tinggi x lebar) berdampak pada seberapa banyak pixel yang ada dalam gambar. Semakin tinggi resolusi, semakin banyak pixel yang digunakan untuk merepresentasikan gambar, yang pada gilirannya dapat menghasilkan gambar yang lebih detail dan halus. Gambar dengan resolusi tinggi memiliki lebih banyak pixel per unit area sehingga gambar yang ditampilakn lebih detail dan halus. Sedangkan gambar dengan resolusi rendah memiliki sedikit pixel per unit area sehingga gambar yang dihasilkan dapat tampak buram dan kurang tajam.

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis vertical ditengah image
for x in range(tinggi):
  image_3[x, int(lebar/2)] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)

#membuat garis menyilang diagonal image
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (tinggi):
  image_3[y,int(y*(lebar/tinggi))] = [255,255,255]
  image_3[y,lebar-1-int(y*(lebar/tinggi))] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
#membuat horizontal putih di tengah gambar
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (100,280):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
#membuat kotak
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (150,280):
  for x in range (100,200):
    image_3[x,y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
